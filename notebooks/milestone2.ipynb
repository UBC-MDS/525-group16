**Reading in Model Data From S3:**

```python
# reading in model data parquet file from s3 bucket and setting time as index
model_data = pd.read_parquet("s3://mds-s3-margotvore/combined_models.parquet/", storage_options=creds)
model_data.set_index('time', inplace=True)
model_data = model_data[model_data['model'] != 'observed'] #  removing redundant observed data entries
model_data
```

**Concatenating, Filtering, Resampling, and Pivoting Observed and Model Data:**

```python
# Merging the observed data and model data together
filtered_df = pd.concat([model_data, observed_data])
```

```python
filtered_df
```

```python
# Filter data to only include Sydney, Australia, resampling, and pivoting dataframe
filtered_df = filtered_df[filtered_df['lat_min'] <= -33.86]
filtered_df = filtered_df[filtered_df['lat_max'] >= -33.86]
filtered_df = filtered_df[filtered_df['lon_min'] <= 151.21]
filtered_df = filtered_df[filtered_df['lon_max'] >= 151.21]
filtered_df.drop(columns=['lat_min', 'lat_max', 'lon_min', 'lon_max'], inplace=True)
filtered_df.index = pd.to_datetime(filtered_df.index)
filtered_df = filtered_df.groupby('model').resample('1D').mean().reset_index()
filtered_df = filtered_df.pivot(index='time', columns='model', values='rain (mm/day)')
filtered_df
```

**Changing Column Order of Output DataFrame:**

```python
# set 'Observed Data' as last column
col_order = [c for c in filtered_df.columns if c not in ["Observed Data"]] + ["Observed Data"]
filtered_df = filtered_df[col_order]
filtered_df
```
